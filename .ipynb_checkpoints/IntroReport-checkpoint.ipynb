# Summary Report: Technology Licensing Database

## Overview
The Technology Licensing Database was constructed in 2017/2018 for research purposes, focusing on technology licensing within the high-tech sector. The dataset is a compendium of various components, each contributing unique information related to Licensing Technology agreements. The data spans from 2006 to 2016 and has been organized in the file `rs_netsales.dta` in Stata 15 format.

## Components of the Dataset
The dataset consists of several key components:

1. **Licensing Technology Agreements (1,480 records):**
   - Sourced exclusively from SIC industries 28, 38, and 73, representing high-tech industries.
   - Information includes agreement type, activation date, duration, royalty rate, territory, tiered rates, and more.

2. **Augmented RoyaltyStat Agreements (400 records):**
   - A subset of Licensing Technology agreements.
   - Contains the original written agreements in PDF and .txt formats.

3. **CompuStat Data (2009-2016):**
   - Financial data for the 1,480 firms in the RoyaltyStat database.
   - Includes information on assets, cash, cost of goods sold, employees, etc.

4. **Identifiers (CIK, Ticker, CUSIP):**
   - Enables match-merging with other datasets.

5. **SDC Platinum Data:**
   - Matched to the firms in the RoyaltyStat database.

## Key Measures and Variables
The dataset contains an extensive array of measures that capture various aspects of the agreements and the involved parties:

- **Agreement Details:** Information about the activation, additional payments, template usage, scope, duration, renewal, and royalty rates.
- **Pricing Clauses (ap*):** Specific pricing clauses such as advertising fees, bonus payments, commission payments, maximum and minimum royalty rates, etc.
- **Assets and IP Exchanged (ata*):** Indicators of the complexity and scope of what is exchanged.
- **Licensee and Licensor Details:** Comprehensive financial data for both licensees and licensors from 1999 to 2016.
- **Industry Indicators:** Redundant industry classifications using SIC, NAICS, and NACE codes.
- **Territorial and Exclusivity Indicators:** Information about the territory of validity, worldwide licensing, and tiered royalty rates.

## Utility and Applications
The Technology Licensing Database offers rich insights into the licensing agreements within the high-tech sector. Potential applications include:

- **Analyzing Licensing Trends:** Understanding the structure and dynamics of licensing agreements.
- **Predicting Royalty Rates:** Utilizing the variables to model and predict royalty rates.
- **Text Mining:** Analyzing the original written agreements for deeper insights.
- **Merging with Other Datasets:** The inclusion of CIK, Ticker, and CUSIP identifiers allows integration with other financial and business data.

## Conclusion
The Technology Licensing Database is a comprehensive and multi-faceted repository of information related to technology licensing. Its diverse components, detailed measures, and extensive temporal scope make it a valuable resource for research and analysis in the field of high-tech industries. The dataset's complexity and richness also underscore the need for careful handling, thorough understanding, and appropriate methodologies in utilizing it for academic and professional pursuits.
