```python
# Function to read a single text file
def read_txt(file_path):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()
    return text

# Function to read all text files in a directory
def read_all_txt_files(directory):
    file_texts = {}
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            file_id = filename.split('.')[0]
            file_path = os.path.join(directory, filename)
            file_texts[file_id] = read_txt(file_path)
    return file_texts

# Reading the Stata file
df, meta = pyreadstat.read_dta('rs_netsales.dta')

# Further selection and copying
selected_features = [
    'royaltyrate', 'tier', 'sic', 'duration', 'territory',
    'naics1', 'template', 'year', 'worldwide', 'market_gdp', 'ref'  # Adding 'ref' here
]
df_model1 = df[selected_features].copy()
```

```python
def filter_rows_with_text(df):
    """
    Filters out rows where the contract_text column contains None.
    
    Parameters:
        df (pandas.DataFrame): The DataFrame to filter.
        
    Returns:
        pandas.DataFrame: A new DataFrame containing only rows with text data in the contract_text column.
    """
    # Drop rows where 'contract_text' is None
    df_filtered = df.dropna(subset=['contract_text'])
    
    return df_filtered

# Usage example:
df_model1_filtered = filter_rows_with_text(df_model1)
```

```python
%%time

# Generate a Word Cloud
def generate_wordcloud(df, column):
    # Combine all rows of the column into one large string
    long_text = ' '.join(df[column])
    wordcloud = WordCloud(width=800, height=400).generate(long_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud")
    plt.show()

# Generate a Histogram of Document Lengths
def plot_document_lengths(df, column):
    df['doc_length'] = df[column].apply(lambda x: len(x.split()))
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df['doc_length'], bins=50, kde=True)
    plt.title("Document Lengths Histogram")
    plt.xlabel("Document Length")
    plt.ylabel("Frequency")
    plt.show()

# Generate a bar plot of the Top-N frequent words
def plot_top_n_words(df, column, n=20):
    long_text = ' '.join(df[column])
    tokens = long_text.split()
    freq = Counter(tokens)
    common_words = freq.most_common(n)
    
    words = [word[0] for word in common_words]
    counts = [word[1] for word in common_words]
    
    plt.figure(figsize=(10, 5))
    sns.barplot(x=counts, y=words)
    plt.title("Top {} Most Frequent Words".format(n))
    plt.xlabel("Frequency")
    plt.ylabel("Words")
    plt.show()

# Usage example
generate_wordcloud(df_model1_filtered, 'contract_text')
plot_document_lengths(df_model1_filtered, 'contract_text')
plot_top_n_words(df_model1_filtered, 'contract_text')
```

```python
# Download the stopwords from nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

new_stopwords = ["all", "due", "to", "on", "daily","b"]

stop_words.update(new_stopwords)
```

```python


def preprocess_text(text):
    """
    Preprocesses the given text by applying tokenization, lowercasing, stopword and special character removal.
    
    Parameters:
        text (str): The text to preprocess.
        
    Returns:
        str: The preprocessed text.
    """
    # Tokenization and lowercasing
    tokens = nltk.word_tokenize(text.lower())
    
    # Stopword Removal
    tokens = [word for word in tokens if word not in stop_words]
    
    # Special Character Removal
    tokens = [re.sub(r'\W+', '', word) for word in tokens]
    
    # Removing tokens that are now empty after special character removal
    tokens = [word for word in tokens if word]
    
    return ' '.join(tokens)

# Apply text preprocessing to each contract text
df_model1_filtered['processed_text'] = df_model1_filtered['contract_text'].apply(preprocess_text)
```

```python
%%time

# Generate a Word Cloud
def generate_wordcloud(df, column):
    # Combine all rows of the column into one large string
    long_text = ' '.join(df[column])
    wordcloud = WordCloud(width=800, height=400).generate(long_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud")
    plt.show()

# Generate a Histogram of Document Lengths
def plot_document_lengths(df, column):
    df['doc_length'] = df[column].apply(lambda x: len(x.split()))
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df['doc_length'], bins=50, kde=True)
    plt.title("Document Lengths Histogram")
    plt.xlabel("Document Length")
    plt.ylabel("Frequency")
    plt.show()

# Generate a bar plot of the Top-N frequent words
def plot_top_n_words(df, column, n=20):
    long_text = ' '.join(df[column])
    tokens = long_text.split()
    freq = Counter(tokens)
    common_words = freq.most_common(n)
    
    words = [word[0] for word in common_words]
    counts = [word[1] for word in common_words]
    
    plt.figure(figsize=(10, 5))
    sns.barplot(x=counts, y=words)
    plt.title("Top {} Most Frequent Words".format(n))
    plt.xlabel("Frequency")
    plt.ylabel("Words")
    plt.show()

# Usage example
generate_wordcloud(df_model1_filtered, 'processed_text')
plot_document_lengths(df_model1_filtered, 'processed_text')
plot_top_n_words(df_model1_filtered, 'processed_text')
```

```python
%%time

# Function to plot word frequency distribution
def plot_word_frequency(text, num_words=30):
    tokens = text.split()
    freq_dist = FreqDist(tokens)
    plt.figure(figsize=(16, 6))
    freq_dist.plot(num_words, cumulative=False)
    plt.show()

# Function to calculate and print average document length
def average_doc_length(df, column):
    avg_length = df[column].apply(lambda x: len(x.split())).mean()
    print(f"The average document length is: {avg_length:.2f} words.")

# Function to plot document length histogram
def plot_doc_length_histogram(df, column):
    plt.figure(figsize=(12, 6))
    doc_lengths = df[column].apply(lambda x: len(x.split()))
    plt.hist(doc_lengths, bins=50, alpha=0.75, color='b')
    plt.title('Document Length Histogram')
    plt.xlabel('Document Length')
    plt.ylabel('Frequency')
    plt.show()

# Function to plot top N n-grams
def plot_top_ngrams(text, n=2, num_ngrams=20):
    tokens = text.split()
    if len(tokens) < n:
        print("Not enough tokens to create n-grams.")
        return
    
    n_grams_list = list(ngrams(tokens, n))
    if not n_grams_list:
        print("No n-grams generated.")
        return
    
    freq_dist = FreqDist(n_grams_list)
    common_ngrams = freq_dist.most_common(num_ngrams)
    
    if not common_ngrams:
        print("No common n-grams found.")
        return
    
    ngrams, counts = zip(*common_ngrams)
    ngrams = [" ".join(gram) for gram in ngrams]
    
    plt.figure(figsize=(16, 6))
    sns.barplot(list(counts), ngrams)
    plt.title(f'Top {num_ngrams} most frequent {n}-grams')
    plt.show()


# Combine all processed text into a single string
all_processed_text = ' '.join(df_model1_filtered['processed_text'])

# Usage examples
plot_word_frequency(all_processed_text)
average_doc_length(df_model1_filtered, 'processed_text')
plot_doc_length_histogram(df_model1_filtered, 'processed_text')
#plot_top_ngrams(all_processed_text, n=2, num_ngrams=20)
```

```python
%%time

def plot_top_n_grams_alternative(text, n, num_ngrams=20):
    tokens = text.split()
    
    if len(tokens) < n:
        print("Not enough tokens to create n-grams.")
        return
    
    n_grams_list = [' '.join(tokens[i:i+n]) for i in range(len(tokens) - n + 1)]
    counter = Counter(n_grams_list)
    
    common_ngrams = counter.most_common(num_ngrams)
    
    if not common_ngrams:
        print("No common n-grams found.")
        return
    
    ngrams, counts = zip(*common_ngrams)
    
    plt.figure(figsize=(16, 6))
    sns.barplot(x=list(counts), y=list(ngrams))
    plt.title(f'Top {num_ngrams} most frequent {n}-grams')
    plt.show()

# Tried using the alternative function
plot_top_n_grams_alternative(all_processed_text, n=2, num_ngrams=20)
plot_top_n_grams_alternative(all_processed_text, n=3, num_ngrams=20)
plot_top_n_grams_alternative(all_processed_text, n=4, num_ngrams=20)
plot_top_n_grams_alternative(all_processed_text, n=5, num_ngrams=20)
```

# Topic Modelling

```python
# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.9, min_df=2, stop_words='english')

# Fit and transform the processed titles
tfidf = tfidf_vectorizer.fit_transform(df_model1_filtered['processed_text'])

# View shape to verify it's been vectorized correctly
tfidf.shape
```

```python
%%time

# Number of topics
n_topics = 5  

# Initialize LDA model
lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=0)

# Fit the model to TF-IDF features
lda_model.fit(tfidf)
```

```python
%%time

def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx+1}:")
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Number of top words to display for each topic
no_top_words = 10

# Get the TF-IDF feature names
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()

# Display the top words in each topic
display_topics(lda_model, tfidf_feature_names, no_top_words)
```

```python
# Get the topic distribution for each document
doc_topic_dist = lda_model.transform(tfidf)

# Convert to DataFrame for easier handling
doc_topic_df = pd.DataFrame(doc_topic_dist, columns=[f'Topic_{i+1}' for i in range(n_topics)])

# Concatenate with the original DataFrame
df_model1_with_topics = pd.concat([df_model1_filtered.reset_index(drop=True), doc_topic_df], axis=1)
```

```python
def plot_word_cloud(lda_model, n_words=10):
    for i, topic in enumerate(lda_model.components_):
        word_cloud = WordCloud().generate_from_frequencies(
            {word: value for word, value in zip(tfidf_vectorizer.get_feature_names_out(), topic)}
        )
        plt.figure()
        plt.imshow(word_cloud)
        plt.title(f"Topic {i+1}")
        plt.axis("off")

plot_word_cloud(lda_model)
```

```python
def plot_heatmap(lda_model, n_words=10):
    topic_term_distributions = lda_model.components_ / lda_model.components_.sum(axis=1)[:, np.newaxis]
    df = pd.DataFrame(topic_term_distributions.T,
                      index=tfidf_vectorizer.get_feature_names_out(),
                      columns=[f"Topic {i+1}" for i in range(lda_model.n_components)])
    
    plt.figure(figsize=(10, 10))
    sns.heatmap(df.sort_values(by='Topic 1', ascending=False).head(n_words))
    plt.title("Topic-Term Heatmap")
    
plot_heatmap(lda_model)
```

```python
from sklearn.manifold import TSNE

def plot_tsne(lda_model):
    topic_term_distributions = lda_model.components_ / lda_model.components_.sum(axis=1)[:, np.newaxis]
    tsne = TSNE(n_components=2, random_state=42)
    topics_tsne = tsne.fit_transform(topic_term_distributions)
    
    plt.figure(figsize=(10, 10))
    plt.scatter(topics_tsne[:, 0], topics_tsne[:, 1])
    for i, txt in enumerate(range(lda_model.n_components)):
        plt.annotate(txt, (topics_tsne[i, 0], topics_tsne[i, 1]))
    plt.title("Topic Distance via t-SNE")

plot_tsne(lda_model)
```
