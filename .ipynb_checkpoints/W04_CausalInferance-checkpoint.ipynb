## Causal Inference

### Association (aka corr. ) vs Causation

## Creating a simple casual inference model to understand variable relationship

variables of interest = ['royaltyrate', 'tier', 'activatedate', 'sic', 'effectiveyear', 'duration', 'territory','naics1', 'template','year','worldwide','market_gdp']

objective: to find how the variables are related to royalty, but clean data first before doing anything else:

**Variables to Consider**
- 'royaltyrate': Target variable to be predicted.
- 'tier': Important as it directly relates to the royalty rate structure.
- 'activatedate' and 'effectiveyear': Consider using one of them to avoid redundancy. 'effectiveyear' might be more useful since it has no missing values.
- 'sic': Industry information. Keep it as it likely affects the royalty rate.
- 'duration': Important feature; consider handling "evergreen" as a special category or numerical value.
- 'territory' and 'worldwide': Territory information; consider encoding 'worldwide' into a binary variable.
- 'naics1': Additional industry information; consider how it relates to 'sic' and decide whether to include both.
- 'template': Potential categorical variable; consider if it has meaningful impact on royalty rates.
- 'year': Similar to 'effectiveyear'; consider using only one.
- 'market_gdp': Economic variable; could be important to reflect market conditions.

**Reasoning**
- 'year': only using this variable as its effectively the same as effectiveyear, and without missing values unlike activedate
- 'worldwide': already binary
- 'template': keeping it as i do not have enough domain knowledge to know if it has a relatioinship
- 'naics1' and 'sic': again, lack  enough domain knowledge to know how sic and naics1 are related except for the fact that theyre both related to industry
